```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# Load Image
# ------------------------------
image = cv2.imread("vikas.jpg")  # put your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ------------------------------
# Window Coordinates (User Input)

xw_min, yw_min = 0, 0
xw_max, yw_max = 1000, 1000

# Crop Window
window = image[yw_min:yw_max, xw_min:xw_max]


# Viewport Coordinates (User Input)

xv_min, yv_min = 0, 0
xv_max, yv_max = 400, 400

# Compute Width & Height

window_width = xw_max - xw_min
window_height = yw_max - yw_min

viewport_width = xv_max - xv_min
viewport_height = yv_max - yv_min

# Scaling Factors

Sx = viewport_width / window_width
Sy = viewport_height / window_height

print("Scaling Factors:")
print("Sx =", Sx)
print("Sy =", Sy)


# Resize Window to Viewport Size

resized = cv2.resize(window, (viewport_width, viewport_height))


# Create Output Canvas

output = np.zeros((500, 500, 3), dtype=np.uint8)

# Place in viewport
output[yv_min:yv_max, xv_min:xv_max] = resized


# Display Results

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(window)
plt.title("Selected Window")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(output)
plt.title("Viewport Mapping Output")
plt.axis("off")

plt.show()
```
